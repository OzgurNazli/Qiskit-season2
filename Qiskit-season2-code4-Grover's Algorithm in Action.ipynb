```python
# Satisfiability problem -> for Optimization Usage of Grover's Algorithm 
# If I want a nearby restourant which one satisfies of my N constraints
# Coding constraints in Oracle

#         GROVER'S ALGORITHM FOR DINNER PARTY 
```

```python
from qiskit import Aer
from qiskit.algorithms import Grover, AmplificationProblem
```

```python
from qiskit.primitives import Sampler
```

```python
from qiskit.circuit.library import PhaseOracle
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
log_exp = '((Olivia & Abe) | (Jin & Amira)) & ~(Abe & Amira)'
oracle = PhaseOracle(log_exp)
problem = AmplificationProblem(oracle)
backend = Aer.get_backend("aer_simulator")

grover = Grover(sampler=Sampler())
result = grover.amplify(problem)
```

```python
counts = result.circuit_results
plot_histogram(counts, title = "PossibleParty Combinations", bar_labels =True)
```
